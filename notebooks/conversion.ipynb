```python
from PIL import Image, ImageDraw, ImageChops

import matplotlib.pyplot as plt

def hole_fill(image_path, output_path):
    # Open the BMP image
    img = Image.open(image_path)

    # Convert the image to RGB (if it's not already)
    img = img.convert("RGB")

    # Get the size of the image
    width, height = img.size

    # Extract the binary mask from the blue channel (assuming it's a 24-bit BMP with R, G, B channels)
    binary_mask = img.split()[2]

    # Create a copy of the binary mask
    filled_mask = binary_mask.copy()
    plt.imshow(filled_mask)
    plt.show()

    # Perform hole-filling using flood-fill
    plt.imshow(filled_mask)
    plt.show()

    ImageDraw.floodfill(filled_mask, (0, 0), 255)

    plt.imshow(filled_mask)
    plt.show()

    # Invert the filled mask
    #filled_mask = ImageChops.invert(filled_mask)
    #plt.imshow(filled_mask)


    # Combine the original image and the filled mask
    #filled_mask = ImageChops.subtract(filled_mask, binary_mask)
    #plt.imshow(filled_mask)
    #plt.title('subtracted')
    #plt.show()

    #result_img = Image.composite(img, Image.new("RGB", img.size, "white"), filled_mask)
    # invert filled_mask:
    result_img = ImageChops.invert(filled_mask)
    
    plt.imshow(result_img)
    plt.show()

    # Save the result
    #result_img.save(output_path)

    return result_img

if __name__ == "__main__":
    input_image_path = r"C:\Users\koenk\Documents\Master_Thesis\Data\BlastsOnline\GT_ZP\Blast_PCRM_R16-0330A ZP_Mask.bmp"
    output_image_path = "output_image.bmp"
    
    result_img = hole_fill(input_image_path, output_image_path)

```
